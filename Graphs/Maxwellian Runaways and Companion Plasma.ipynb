```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Maxw(v, T=300, m=1, k=1):
    return 4*np.pi*(m / (2*np.pi*k*T))**(3/2) * v**2 * np.e**(-m*v**2 / (2*k*T))
```

```python
x = np.linspace(0, 300, 10000)
plt.plot(x, Maxw(x))
plt.plot(x, Maxw(x, T=4000))
plt.xlabel('v in au')
plt.ylabel('$f(v)$')
plt.savefig('Runaways and companion plasma.png',dpi=369)
```
